```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # Ajout de l'import manquant

# Lecture du fichier
chemin_fichier = "/Users/ludovicveltz/Documents/Bootcamp_GENAI_2025/Crashcourse/hackaton_projet_W3D5/DATASET/plats_populaires_scores_sante_normalises.csv"
df = pd.read_csv(chemin_fichier)

# Configuration du style
plt.style.use('default')
plt.rcParams['figure.figsize'] = (15, 10)

# 1. Comparaison des macronutriments par pays
fig, axes = plt.subplots(2, 2, figsize=(20, 15))
fig.suptitle('Comparaison des valeurs nutritionnelles par pays', fontsize=16)

# Calories
sns.boxplot(data=df, x='Pays', y='Calories (kcal)', ax=axes[0,0])
axes[0,0].set_title('Distribution des calories par pays')
axes[0,0].tick_params(axis='x', rotation=45)

# Protéines
sns.boxplot(data=df, x='Pays', y='Protéines (g)', ax=axes[0,1])
axes[0,1].set_title('Distribution des protéines par pays')
axes[0,1].tick_params(axis='x', rotation=45)

# Lipides
sns.boxplot(data=df, x='Pays', y='Lipides (g)', ax=axes[1,0])
axes[1,0].set_title('Distribution des lipides par pays')
axes[1,0].tick_params(axis='x', rotation=45)

# Glucides
sns.boxplot(data=df, x='Pays', y='Glucides (g)', ax=axes[1,1])
axes[1,1].set_title('Distribution des glucides par pays')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# 2. Analyse des ratios nutritionnels par pays
fig, axes = plt.subplots(2, 1, figsize=(15, 12))

# Configuration du style avant la création du graphique
plt.style.use('default')  # Utilisation du style par défaut de matplotlib

# Ratio Protéines/Lipides
df['Ratio_Prot_Lip'] = df['Protéines (g)'] / df['Lipides (g)']

# Création du boxplot avec échelle ajustée
sns.boxplot(data=df, x='Pays', y='Ratio_Prot_Lip', ax=axes[0])
axes[0].set_title('Ratio Protéines/Lipides par pays', pad=20, fontsize=12)
axes[0].tick_params(axis='x', rotation=45)
axes[0].set_ylim(0, 5)

# Ajout d'une grille horizontale pour meilleure lisibilité
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Personnalisation des labels
axes[0].set_xlabel('Pays', fontsize=10)
axes[0].set_ylabel('Ratio Protéines/Lipides', fontsize=10)

# Ajustement de la mise en page
plt.tight_layout()



# Score Santé
sns.boxplot(data=df, x='Pays', y='Score_Sante', ax=axes[1])
axes[1].set_title('Score Santé par pays')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# 3. Analyse statistique comparative
stats_pays = df.groupby('Pays').agg({
    'Calories (kcal)': ['mean', 'std'],
    'Protéines (g)': ['mean', 'std'],
    'Lipides (g)': ['mean', 'std'],
    'Glucides (g)': ['mean', 'std'],
    'Score_Sante': ['mean', 'std']
}).round(2)

print("\nAnalyse nutritionnelle comparative par pays:")
print(stats_pays)

# 4. Test statistique pour comparer les moyennes entre pays
from scipy import stats

print("\nTests statistiques des différences entre pays:")
nutriments = ['Calories (kcal)', 'Protéines (g)', 'Lipides (g)', 'Glucides (g)', 'Score_Sante']

for nutriment in nutriments:
    # ANOVA à un facteur
    groupes = [group[nutriment].values for name, group in df.groupby('Pays')]
    f_stat, p_val = stats.f_oneway(*groupes)
    
    print(f"\n{nutriment}:")
    print(f"p-value: {p_val:.4f}")
    if p_val < 0.05:
        print("→ Différences significatives entre les pays")
    else:
        print("→ Pas de différence significative entre les pays")

# 5. Identification des plats extrêmes par pays
print("\nPlats caractéristiques par pays:")
for pays in df['Pays'].unique():
    plats_pays = df[df['Pays'] == pays]
    print(f"\n{pays}:")
    print(f"Plat le plus calorique: {plats_pays.loc[plats_pays['Calories (kcal)'].idxmax(), 'Plat']} "
          f"({plats_pays['Calories (kcal)'].max():.0f} kcal)")
    print(f"Plat le plus sain: {plats_pays.loc[plats_pays['Score_Sante'].idxmax(), 'Plat']} "
          f"(Score: {plats_pays['Score_Sante'].max():.1f})")

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Lecture des fichiers
disponibilite = pd.read_csv("DATASET/disponibilite_ingredients.csv", index_col=0)
popularite = pd.read_csv("DATASET/plats_populaires_mondial.csv")

# Liste des 10 pays
pays = ['France', 'Chine', 'Inde', 'Italie', 'Japon', 'Mexique', 'Espagne', 'Thaïlande', 'USA', 'Grèce']

# Définition des groupes avec uniquement les ingrédients disponibles
ingredients_groupes = {
    'Viandes et Volailles': ['chicken', 'chicken breast', 'chicken legs', 'chicken thighs', 'chicken wings'],
    'Fruits de Mer': ['salmon', 'prawns', 'clams', 'shrimp', 'mussels', 'lobster', 'octopus'],
    'Fromages et Laitages': ['cheese', 'mozzarella', 'parmesan', 'feta'],
    'Céréales et Pâtes': ['rice', 'pasta', 'noodles']
}

# 1. Analyse par groupe d'ingrédients
plt.figure(figsize=(15, 8))
for i, (groupe, ingredients) in enumerate(ingredients_groupes.items()):
    moyennes = disponibilite.loc[ingredients][pays].mean()
    plt.bar([x + i*0.2 for x in range(len(pays))], moyennes, width=0.2, 
            label=groupe, alpha=0.7)

plt.title('Disponibilité moyenne par groupe d\'ingrédients et par pays')
plt.xticks([x + 0.3 for x in range(len(pays))], pays, rotation=45)
plt.ylabel('Score de disponibilité (1-10)')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

# 2. Heatmap des disponibilités par groupe
disponibilite_groupes = pd.DataFrame()
for groupe, ingredients in ingredients_groupes.items():
    disponibilite_groupes[groupe] = disponibilite.loc[ingredients][pays].mean()

plt.figure(figsize=(12, 8))
sns.heatmap(disponibilite_groupes.T, annot=True, fmt='.1f', cmap='YlOrRd')
plt.title('Disponibilité moyenne par groupe d\'ingrédients et pays')
plt.tight_layout()
plt.show()

# 3a. Analyse statistique basique
print("\nAnalyse par groupe d'ingrédients:")
for groupe, ingredients in ingredients_groupes.items():
    ingredients_disponibles = [ing for ing in ingredients if ing in disponibilite.index]
    if ingredients_disponibles:
        moyennes = disponibilite.loc[ingredients_disponibles][pays].mean().mean()
        variation = disponibilite.loc[ingredients_disponibles][pays].std().mean()
        print(f"\n{groupe}:")
        print(f"Disponibilité moyenne globale: {moyennes:.2f}")
        print(f"Variation moyenne: {variation:.2f}")
        meilleur_pays = disponibilite.loc[ingredients_disponibles][pays].mean().idxmax()
        print(f"Meilleure disponibilité: {meilleur_pays}")

# 3b. Analyse statistique détaillée
print("\nAnalyse détaillée par groupe d'ingrédients:")
for groupe, ingredients in ingredients_groupes.items():
    moyennes = disponibilite.loc[ingredients][pays].mean()
    print(f"\n{groupe}:")
    print(f"Top 3 pays avec la meilleure disponibilité:")
    print(moyennes.nlargest(3))
    print(f"Disponibilité moyenne globale: {moyennes.mean():.2f}")
    print(f"Écart type: {moyennes.std():.2f}")

# 4. Corrélation entre groupes d'ingrédients
correlations = pd.DataFrame()
for groupe, ingredients in ingredients_groupes.items():
    correlations[groupe] = disponibilite.loc[ingredients][pays].mean()

plt.figure(figsize=(10, 8))
sns.heatmap(correlations.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Corrélation entre les groupes d\'ingrédients')
plt.tight_layout()
plt.show()
```
