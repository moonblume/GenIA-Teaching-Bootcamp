```python
print("\n========== CONFIGURATION INITIALE ==========")
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

API_KEY = "V47bTvrwdE63sv3FPzRPbyo30fxLmg4cJnc6tWol"
base_url = "https://api.nal.usda.gov/fdc/v1/foods/search"

print("\n========== FONCTION DE REQUÊTE API ==========")
def get_food_data(query):
    """
    Fonction pour récupérer les données d'un aliment
    """
    params = {
        "api_key": API_KEY,
        "query": query,
        "pageSize": 5  # Limitons à 5 résultats pour l'exemple
    }
    try:
        response = requests.get(base_url, params=params)
        return response.json()
    except Exception as e:
        print(f"Erreur lors de la requête : {e}")
        return None
```

```python
def afficher_liste_aliments():
    """
    Fonction pour afficher un échantillon d'aliments et leurs données nutritionnelles
    Utilisation de l'approche NumPy pour le traitement des données
    """
    # Liste d'aliments communs pour l'exemple
    aliments_test = ["apple", "bread", "milk", "egg", "rice"]
    
    print("=== ANALYSE DES DONNÉES NUTRITIONNELLES ===\n")
    
    for aliment in aliments_test:
        print(f"\nAnalyse de : {aliment.upper()}")
        print("-" * 50)
        
        # Récupération des données
        data = get_food_data(aliment)
        
        if data and 'foods' in data:
            for i, food in enumerate(data['foods'], 1):
                print(f"\nVariété {i}:")
                print(f"Nom: {food.get('description', 'Non spécifié')}")
                print(f"Catégorie: {food.get('foodCategory', 'Non spécifié')}")
                
                # Analyse des nutriments
                if 'foodNutrients' in food:
                    print("\nNutriments principaux:")
                    for nutrient in food['foodNutrients']:
                        if nutrient.get('nutrientName'):
                            print(f"- {nutrient['nutrientName']}: "
                                  f"{nutrient.get('value', 'N/A')} "
                                  f"{nutrient.get('unitName', '')}")
                
                print("\nInformations additionnelles:")
                print(f"Marque: {food.get('brandOwner', 'Non spécifié')}")
                print(f"Portion: {food.get('servingSize', 'Non spécifié')} "
                      f"{food.get('servingSizeUnit', '')}")
        else:
            print("Données non disponibles pour cet aliment")
            
        print("\n" + "="*50)

# Exécution de l'analyse
print("Démarrage de l'analyse des aliments...")
afficher_liste_aliments()
```
