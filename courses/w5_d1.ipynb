```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

# Télécharger les ressources nécessaires
nltk.download('punkt')
nltk.download('stopwords')

# Texte à traiter
doc = '''Linguistics and Natural Language Processing (NLP) are closely linked. \n Linguistics is the scientific study of language, encompassing its structure, meaning, and context. \n It provides foundational knowledge about language syntax, semantics, pragmatics, and phonetics. \n NLP applies this linguistic knowledge in computational algorithms to enable computers to understand, interpret, and generate human language. By leveraging linguistic principles, NLP seeks to bridge the gap between human communication and computer understanding, enabling tasks like translation, sentiment analysis, and voice recognition.'''

# Nettoyer le texte en supprimant les caractères non alphabétiques
doc_cleaned = re.sub(r'\W+', ' ', doc)

# Tokeniser le texte
tokens = word_tokenize(doc_cleaned)

# Filtrer les mots vides
filtered_words = [word for word in tokens if word.lower() not in stopwords.words('english')]

# Afficher les tokens originaux et filtrés
print("Original Tokens:", tokens)
print("Filtered Tokens:", filtered_words)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk import pos_tag
import spacy

# Vérifier le chemin d'accès aux données NLTK
print("NLTK data path:", nltk.data.path)

# Télécharger les ressources nécessaires
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng', force=True)
nltk.download('tagsets_json')

# Texte à traiter
doc_text = '''Charles Lutwidge Dodgson, better known by his pen name Lewis Carroll, was an English author, poet, mathematician and photographer. His most notable works are Alice's Adventures in Wonderland (1865) and its sequel Through the Looking-Glass (1871).'''

# Tokenisation avec NLTK
tokens = word_tokenize(doc_text)

# Initialiser le stemmer
stemmer = PorterStemmer()

# Appliquer le stemming aux tokens
stemmed = [stemmer.stem(word) for word in tokens]
print('Stemmed:', stemmed)

# Charger le modèle de langue anglaise de spaCy
nlp = spacy.load('en_core_web_sm')

# Traiter le texte avec spaCy
doc = nlp(doc_text)

# Filtrer les mots vides et les caractères non alphabétiques
stop_words = set(stopwords.words('english'))
filtered_doc = [token for token in doc if token.text.lower() not in stop_words and token.is_alpha]

# Appliquer la lemmatisation
lemmatized = [token.lemma_ for token in filtered_doc]
print('Lemmatized:', lemmatized)

# Appliquer le POS tagging avec NLTK
tagged = pos_tag(tokens)
print("\nPart of Speech Tags:")
for word, tag in tagged:
    print(f'Word: {word}, POS: {tag}')

# Vérifier la signification des tags
nltk.help.upenn_tagset('NN')
```

```python
import json
from sklearn.feature_extraction.text import TfidfVectorizer

with open('famous_poems.json', 'r') as f:
    data = json.load(f)

processed_poems = []
for poem in data:
    tokens = word_tokenize(poem['text'])
    filtered_tokens = [word for word in tokens if not word.lower() in stopwords.words('english') and word.isalpha()]
    processed_poems.append(' '.join(filtered_tokens))

# Apply TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(processed_poems)

#Bonus:
# Find the most important word and print
feature_names = vectorizer.get_feature_names_out()
for i, row in enumerate(tfidf_matrix.toarray()):
    max_index = row.argmax()
    most_important_word = feature_names[max_index]
    print(f"Most important word in poem '{data[i]['title']}': {most_important_word}")
```

**Comment word2vec fonctionne ?**
Il existe deux variantes du Word2vec, les deux utilisent un réseau de neurones à 3 couches (1 couche d’entrée, 1 couche cachée, 1 couche de sortie) : Common Bag Of Words (CBOW) et Skip-gram.

CBOW : Le modèle est nourri par le contexte, et prédit le mot cible. Le résultat de la couche cachée est la nouvelle représentation du mot (ℎ1, …, ℎ𝑁).

Skip Gram : Le modèle est nourri par le mot cible, et prédit les mots du contexte. Le résultat de la couche cachée est la nouvelle représentation du mot (ℎ1, …, ℎ𝑁).

```python
# Liste de messages à classifier
messages = [
    "Congratulations! You have won a free ticket to Bahamas.",
    "Please review the project proposal and send your feedback.",
    "Limited time offer! Get a free trial now.",
    "Let's schedule a meeting to discuss the project details.",
    "Win a brand new car by entering our competition."
]

# Mots-clés associés au spam
spam_keywords = {"win", "free", "offer"}

# Mots-clés associés à des messages légitimes
not_spam_keywords = {"meeting", "project"}
```

```python
# Fonction pour classifier un message
def classify_message(message):
    # Convertir le message en minuscules et le diviser en mots
    words = set(message.lower().split())

    # Vérifier la présence de mots-clés de spam
    if words & spam_keywords:
        return "Spam"
    # Vérifier la présence de mots-clés de non-spam
    elif words & not_spam_keywords:
        return "Not Spam"
    else:
        return "Not Spam"

# Classifier chaque message
classified_messages = [(message, classify_message(message)) for message in messages]

# Afficher les résultats
for message, classification in classified_messages:
    print(f"Message: '{message}' -> {classification}")
```

Load and Preprocess the Data

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
df = pd.read_csv("spam.csv", encoding="latin-1")
df = df[["v1", "v2"]]  # Keep only the label and text columns
df.columns = ["label", "text"]

# Convert labels to binary (spam = 1, ham = 0)
df["label"] = df["label"].map({"spam": 1, "ham": 0})

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df["text"], df["label"], test_size=0.2, random_state=42)
```

Feature extaction with TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert text to TF-IDF features
vectorizer = TfidfVectorizer(stop_words="english")
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

Train a Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Train the model
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Make predictions
y_pred = model.predict(X_test_tfidf)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

Le rapport de classification que vous avez fourni évalue la performance d'un modèle de classification binaire, probablement pour distinguer entre "Spam" et "Not Spam". Voici un résumé simplifié :

- **Exactitude (Accuracy)** : Le modèle a correctement classé environ 97% des messages. Cela signifie qu'il est généralement bon pour distinguer entre les deux catégories.

- **Classe "Not Spam"** :
  - Le modèle est très précis (96%) et a capturé tous les exemples de cette classe (rappel de 100%). Cela signifie qu'il identifie presque toujours correctement les messages qui ne sont pas du spam.

- **Classe "Spam"** :
  - Le modèle est parfait en précision (100%), ce qui signifie qu'il ne classe pas à tort les messages non-spam comme spam.
  - Cependant, il a un rappel de 75%, ce qui signifie qu'il manque certains messages qui sont réellement du spam.

- **Conclusion** : Le modèle fonctionne bien dans l'ensemble, mais il pourrait être amélioré pour mieux capturer tous les messages de spam sans en manquer. Cela pourrait être important si l'objectif est de minimiser les faux négatifs (messages de spam non détectés).
