**Comparison of Traditional Machine Learning vs. Deep Learning**

**Traditional Machine Learning/Deep Learning**

**Feature Engineering**
Requires manual feature extraction and selection/
Learns features automatically from data

**Data Processing**

Performs well with structured and smaller datasets/
Requires large amounts of data, often unstructured

**Scalability**

Struggles with high-dimensional and large-scale data/
Highly scalable with large datasets and complex problems

**Pattern Discovery**

Detects patterns based on pre-defined features/
Learns hierarchical patterns and complex representations

**Computational Requirements**

Generally requires less computational power	/
Requires high computational resources (e.g., GPUs, TPUs)


**Real-World Problem Examples**
**Traditional ML is better suited for: **
Credit scoring – Traditional algorithms like logistic regression or decision trees can efficiently analyze structured tabular data with predefined features.
**Deep Learning is superior for: **
Autonomous driving – Deep neural networks, especially convolutional neural networks (CNNs), can process vast amounts of image and sensor data to recognize objects and make driving decisions.


**Deep Learning and Unstructured Data**


Deep learning excels in processing unstructured data—such as images, text, and audio—because it can automatically extract hierarchical features without human intervention. Unlike traditional ML, which relies on manual feature engineering, deep neural networks learn directly from raw data, identifying complex patterns through layers of abstraction. This ability makes deep learning highly effective in tasks like natural language processing (NLP), computer vision, and speech recognition, where unstructured data dominates.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

np.random.seed(0)
x = np.arange(-1, 1, 0.1)
y = -x**2 + np.random.normal(0, 0.05, len(x))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Generated Data', color='blue', marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of Generated Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_train = x[:12]
y_train = y[:12]
x_test = x[12:]
y_test = y[12:]

print("x_train:", x_train)
print("y_train:", y_train)
print("x_test:", x_test)
print("y_test:", y_test)
```

```python
def polynomial_fit(degree):
  coefficients = np.polyfit(x_train, y_train, degree)
  polynomial = np.poly1d(coefficients)
  return polynomial
```

```python
def plot_polyfit(degree):
    polynomial = polynomial_fit(degree)
    y_train_pred = polynomial(x_train)
    y_test_pred = polynomial(x_test)

    plt.figure(figsize=(8, 6))
    plt.scatter(x_train, y_train, label='Training Data', color='blue', marker='o')
    plt.scatter(x_test, y_test, label='Test Data', color='red', marker='x')
    plt.plot(x, polynomial(x), label=f'Polynomial Fit (degree={degree})', color='green')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Polynomial Fit of Degree {degree}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_polyfit(1)
```

```python
plot_polyfit(7)
```

```python
plot_polyfit(11)
```

```python
for degree in range(1, 12):
    plot_polyfit(degree)
```

```python
def calculate_rmse(degree):
    polynomial = polynomial_fit(degree)
    y_train_pred = polynomial(x_train)
    y_test_pred = polynomial(x_test)

    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

    return rmse_train, rmse_test

for degree in range(1, 12):
    rmse_train, rmse_test = calculate_rmse(degree)
    print(f"Degree {degree}: RMSE (Training) = {rmse_train:.4f}, RMSE (Test) = {rmse_test:.4f}")
```

```python
results = []
for degree in range(1, 12):
    rmse_train, rmse_test = calculate_rmse(degree)
    results.append([degree, rmse_train, rmse_test])

results
```

```python
import matplotlib.pyplot as plt

degrees = [result[0] for result in results]
rmse_train_values = [result[1] for result in results]
rmse_test_values = [result[2] for result in results]

plt.figure(figsize=(10, 6))
plt.plot(degrees, rmse_train_values, marker='o', label='Training RMSE')
plt.plot(degrees, rmse_test_values, marker='x', label='Test RMSE')
plt.xlabel('Polynomial Degree')
plt.ylabel('RMSE (Log Scale)')
plt.yscale('log')
plt.title('RMSE vs. Polynomial Degree')
plt.legend()
plt.grid(True)
plt.show()
```

```python
min_rmse_test = min(results, key=lambda x: x[2])
min_rmse_degree = min_rmse_test[0]
min_rmse_test_value = min_rmse_test[2]

print(f"The degree that minimizes the RMSE on the test set is {min_rmse_degree} with a RMSE of {min_rmse_test_value:.4f}")

if min_rmse_degree == 2:
    print("This corresponds to the true underlying model (y = -x^2)")
else:
    print("This does not correspond to the true underlying model (y = -x^2)")
```
